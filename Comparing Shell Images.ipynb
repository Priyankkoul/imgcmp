```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

    
    
labels= [23,67,56,12,16]

also_labels= labels

#Correlation Coefficients
testing_array = np.array([[1,0.98,0.758,0.856,0.257],
                    [0.98, 1, 0.257, 0.786, 0.854],
                    [0.758, 0.257, 1, 0.657, 0.185],
                    [0.856, 0.786, 0.657, 1, 0.987],
                    [0.257, 0.854, 0.185, 0.987, 1]])


fig, ax = plt.subplots()
im = ax.imshow(testing_array)


ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(also_labels)))



ax.set_xticklabels(labels)
ax.set_yticklabels(also_labels)

# Rotate the tick labels and set their alignment.

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.

for i in range(len(labels)):
    for j in range(len(also_labels)):
        text = ax.text(j, i, testing_array[i, j],
                       ha="center", va="center", color="w")

ax.set_title("This is just Priyank testing something.\n")
fig.tight_layout()
plt.show()

```
